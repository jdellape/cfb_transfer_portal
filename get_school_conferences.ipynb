```python
from __future__ import print_function
import time
import cfbd
from cfbd.rest import ApiException
from pprint import pprint

```

```python
api_instance = cfbd.TeamsApi()

#I think this api uses all 3 letter abbreviations for conferences. Need to replace list with those
#I'm still missing some, but this will probably get me to about what I need.
conference_list = ['ACC', 'B12', 'B1G', 'SEC', 'PAC', 'CUSA',
                  'MAC','MWC','FBSI', 'SBC', 'Ind'
                  'AAC','Western','MVIAA','RMC',
                  'SWC', 'SWAC']

conferences = []
schools = []

for conference in conference_list:
    try:
        # Get School & Conference information
        team_list = api_instance.get_teams(conference=conference)
        for team in team_list:
            conferences.append(team.conference)
            schools.append(team.school)
    except ApiException as e:
        print("Exception when calling TeamsApi->get_teams: %s\n" % e)
```

```python
for school, conference in zip(schools, conferences):
    print(school, conference)
```

```python
#Throw lists into pandas dataframe and download into csv
import pandas as pd
data_dict = {'SCHOOL_NAME': schools, 'CONFERENCE_NAME': conferences}

df = pd.DataFrame(data_dict) 

df.head(10)
```

```python
df.to_csv('schools_with_conferences.csv')
```
